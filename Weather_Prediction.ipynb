```python
from sklearn.linear_model import Ridge

reg=Ridge(alpha=.1)
predictors=["precip","temp_max","temp_min"]
train=core_weather.loc[:"2020-12-31"]
test=core_weather.loc["2021-01-01":] #train data must always be previous data
reg.fit(train[predictors],train["target"])
predictions=reg.predict(test[predictors])

```

```python
from sklearn.metrics import mean_absolute_error
mean_absolute_error(test["target"],predictions)
```

```python
combined=pd.concat([test["target"],pd.Series(predictions, index=test.index)],axis=1)
combined.columns=["actual","predictions"]
combined
```

```python
def create_predictions(predictors,core_weather,reg):
  train=core_weather.loc[:"2020-12-31"]
  test=core_weather.loc["2021-01-01":] #train data must always be previous data
  reg.fit(train[predictors],train["target"])
  predictions=reg.predict(test[predictors])
  error=mean_absolute_error(["target"],predictions)
  combined=pd.concat([test["target"],pd.Series(predictions, index=test.index)],axis=1)
  combined.columns=["actual","predictions"]
  return error,combined
```

```python
core_weather.index = pd.to_datetime(core_weather.index)

# Group by month and calculate the expanding mean
monthly_avg = core_weather["temp_max"].groupby(core_weather.index.month).apply(lambda x: x.expanding(1).mean())

# Assign it back to the DataFrame
core_weather["monthly_avg"] = monthly_avg.values

core_weather
```

```python
# Ensure core_weather has a DatetimeIndex
core_weather.index = pd.to_datetime(core_weather.index)

# Group by the day of the year and calculate the expanding mean
day_of_year_avg = core_weather["temp_max"].groupby(core_weather.index.day_of_year).apply(lambda x: x.expanding(1).mean())

# Assign the result back to the DataFrame using .values to align indices
core_weather["day_of_year_avg"] = day_of_year_avg.values
```

```python
# prompt: write the code to print the error, find the new error

predictors = ["precip", "temp_max", "temp_min", "month_max", "month_min", "month_day_max", "max_min", "monthly_avg", "day_of_year_avg"]

# Ensure core_weather has a DatetimeIndex
core_weather.index = pd.to_datetime(core_weather.index)

# Ensure 'target' is in core_weather, if it's not calculate it again
if 'target' not in core_weather:
    core_weather["target"] = core_weather["temp_max"].shift(-1)
    core_weather = core_weather.iloc[:-1, :].copy()

# Redefine train and test sets based on updated core_weather
train = core_weather.loc[:"2020-12-31"]
test = core_weather.loc["2021-01-01":]

# Re-train the model with the updated predictors
reg = Ridge(alpha=0.1)
reg.fit(train[predictors], train["target"])
predictions = reg.predict(test[predictors])
new_error = mean_absolute_error(test["target"], predictions)

print("New Error:", new_error)
```
